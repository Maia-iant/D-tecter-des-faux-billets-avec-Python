### 2 EDA

```python
def analyse_exploratoire(df, nom_dataset="Dataset"):
    print(f"\n==== Analyse Exploratoire : {nom_dataset} ====\n")
    
    # Dimensions
    print("Dimensions :")
    print(df.shape)
    
    # Types des colonnes et valeurs manquantes
    print("\nTypes et valeurs manquantes :")
    print(df.dtypes)
    print("\nAperçu :")
    display(df.head())
    print("\nPourcentage de valeurs manquantes :\n", df.isnull().mean() * 100)
    
    # Unicité pour les colonnes object (s'il y en a)
    cat_cols = df.select_dtypes(include='object').columns
    if len(cat_cols) > 0:
        print("\nUnicité des variables catégorielles :")
        for col in cat_cols:
            print(f"Valeurs uniques pour {col} :", df[col].nunique())
    else:
        print("\nAucune variable catégorielle (type object).")
        
    # Statistiques descriptives numériques
    print("\nStats descriptives numériques :")
    if df.select_dtypes(include=np.number).shape[1] > 0:
        display(df.describe().T)
    else:
        print("Aucune variable numérique.")
    
    # Statistiques descriptives catégorielles (uniquement si object)
    if len(cat_cols) > 0:
        print("\nStats descriptives catégorielles :")
        display(df[cat_cols].describe(include=['object']).T)
    else:
        print("\nPas de stats descriptives catégorielles : pas de colonnes object.")
    
    # Visualisation valeurs manquantes
    print("\nVisualisation des valeurs manquantes :")
    msno.matrix(df)
    plt.show()
    msno.heatmap(df)
    plt.show()

    # Outliers (boxplots) pour variables numériques
    num_cols = df.select_dtypes(include='number').columns
    for col in num_cols:
        series = df[col].dropna()
        if len(series) > 0:
            plt.figure()
            sns.boxplot(x=series)
            plt.title(f'Boîte à moustaches - {col}')
            plt.show()
    
    # Matrice de corrélation
    num_df = df.select_dtypes(include=np.number)
    print("\nMatrice de corrélation :")
    if num_df.shape[1] == 0:
        print("Pas de colonnes numériques disponibles pour la corrélation.")
    else:
        corr = num_df.corr()
        plt.figure(figsize=(8, 6))
        sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm")
        plt.title("Matrice de Corrélation")
        plt.show()

    # Table de contingence & test Chi2 pour colonnes catégorielles (si au moins 2)
    if len(cat_cols) >= 2:
        print("\nTest du Chi2 sur 2 variables catégorielles :")
        table = pd.crosstab(df[cat_cols[0]], df[cat_cols[1]])
        chi2, p, dof, expected = chi2_contingency(table)
        print(f"Test du Chi2 entre {cat_cols[0]} et {cat_cols[1]} :")
        print("Statistique chi2 :", chi2, "p-value :", p)
        display(table)
    else:
        print("\nPas assez de variables catégorielles pour un test du Chi2.")
    
    # Profilage automatique ydata_profiling
    print("\nGénération du rapport de profilage (ydata_profiling) :")
    try:
        profile = ProfileReport(df, title=f"Profilage {nom_dataset}", explorative=True)
        profile.to_file(f"profil_{nom_dataset}.html")
        print(f"Rapport sauvegardé : profil_{nom_dataset}.html")
    except Exception as e:
        print("Profilage automatique indisponible :", e)
    
    print("\n" + "=" * 60)
```

### 3 Prétraitement

```python
#Définition de x et y
# Cible
y = df["is_genuine"]          # True / False

# Variables explicatives (toutes les colonnes sauf la cible)
X = df.drop(columns=["is_genuine"])
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,#20% des données en test-80% en train
    random_state=42,
    stratify=y#permet de garder la proportion 1000 vrais/500 faux dans chaque échantillon
)
```

### 4-4 Kmeans

### Apprentissage non supervisé

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
#Variables explicatives sans la cible
X = df.drop(columns=["is_genuine"])
```

```python
#Imputation des NaN (par la médiane par exemple)
imputer = SimpleImputer(strategy="median")
X_imputed = imputer.fit_transform(X)
```

```python
#Standardisation
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_imputed)
```

```python
#ACP
pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X_scaled)

print("Variance expliquée :", pca.explained_variance_ratio_)
```

### Les deux premières composantes principales expliquent environ 60% de la variance totale des données, ce qui permet une visualisation pertinente des individus et des centroïdes sur le plan factoriel (PC1, PC2)

```python
kmeans = KMeans(n_clusters=2, random_state=42, n_init=10)
clusters = kmeans.fit_predict(X_pca)

centroids = kmeans.cluster_centers_
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
# Les individus
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=clusters, cmap="viridis", alpha=0.6, label="Individus")

# Les centroïdes
plt.scatter(centroids[:, 0], centroids[:, 1],
            c="red", s=200, marker="X", label="Centroïdes")

plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("K-means sur l'espace ACP (PC1 vs PC2)")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X_pca[:, 0], X_pca[:, 1],
            c=y.map({True:1, False:0}), cmap="coolwarm", alpha=0.6)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("Projection ACP colorée par is_genuine")
plt.show()
```

##### Les deux premières composantes principales expliquent environ 60 % de la variance totale des données.  
##### La figure 1 montre les clusters obtenus par K-means dans le plan factoriel (PC1, PC2) ainsi que leurs centroïdes.  
##### La figure 2 représente la même projection ACP, colorée cette fois par la variable cible `is_genuine`.  
##### On observe que la séparation obtenue par K-means recoupe très bien la distinction entre vrais et faux billets, ce qui confirme la pertinence du clustering dans l’espace réduit par l’ACP.

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2, random_state=42, n_init=10)
clusters = kmeans.fit_predict(X_pca)  # X_pca issu de l'ACP
```

```python
#Trouver la correspondance de chaque cluster
import numpy as np
import pandas as pd

# y est la série is_genuine (True/False)
df_clusters = pd.DataFrame({"cluster": clusters, "is_genuine": y})

majority_mapping = {}
for c in np.unique(clusters):
    # proportion de True dans le cluster c
    prop_true = df_clusters[df_clusters["cluster"] == c]["is_genuine"].mean()
    # si > 0.5, on dit que ce cluster = True, sinon = False
    majority_mapping[c] = True if prop_true >= 0.5 else False

majority_mapping
# ex: {0: False, 1: True}
```

```python
y_pred_kmeans = pd.Series(clusters).map(majority_mapping)
```

### 5- Comparaison des modèles

```python
from sklearn.metrics import accuracy_score, confusion_matrix

def eval_modele(nom, y_true, y_pred):
    acc = accuracy_score(y_true, y_pred)
    cm = confusion_matrix(y_true, y_pred)
    tn, fp, fn, tp = cm.ravel()  # ordre: [[tn, fp],[fn, tp]]
    
    return {
        "Modèle": nom,
        "Accuracy": acc,
        "Faux billets bien détectés": tn,
        "Faux billets mal détectés (FP)": fp,
        "Vrais billets mal détectés (FN)": fn,
        "Vrais billets bien détectés": tp
    }
```

```python
result_kmeans = eval_modele("K-means (centroïdes)", y, y_pred_kmeans)
```

```python
resultats = []
resultats.append(eval_modele("Régression logistique", y_test, y_pred_lr))
resultats.append(eval_modele("KNN", y_test, y_pred_knn))
resultats.append(eval_modele("Random Forest", y_test, y_pred_rf))
resultats.append(result_kmeans)

df_resultats = pd.DataFrame(resultats)
display(df_resultats)
```

#### Les quatre modèles testés (régression logistique, KNN, Random Forest et K-means) présentent des performances très élevées, avec une accuracy comprise entre environ 98,3% et 99%. Cependant, la régression logistique obtient la meilleure performance globale (99% d’accuracy) tout en limitant le nombre de faux billets prédits à tort comme vrais à seulement 2 cas sur 300 billets de test, ce qui répond directement à la priorité de l’ONCFM de réduire au maximum ce type d’erreur. La Random Forest et le KNN offrent des résultats proches, mais avec davantage d’erreurs sur la classe “faux billet”, ce qui les rend moins adaptés comme modèle principal pour l’application.

#### Le modèle non supervisé K-means, appliqué sur les composantes principales issues de l’ACP, atteint une accuracy d’environ 98%.
#### Sur 500 faux billets, 488 sont correctement détectés et 12 sont classés à tort comme vrais (FP).
#### Sur 1000 vrais billets, 983 sont correctement détectés, tandis que 17 sont classés à tort comme faux (FN).
#### Les performances restent légèrement inférieures à celles des modèles supervisés (régression logistique, KNN, Random Forest), mais montrent qu’un simple clustering non supervisé permet déjà de retrouver la structure vraie/faux billets de manière satisfaisante.

### 6 - Création du pipeline complet 

```python
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

pipeline_lr = make_pipeline(
    SimpleImputer(strategy="median"),
    StandardScaler(),
    LogisticRegression(max_iter=1000, random_state=42)
)

pipeline_lr.fit(X_train, y_train)
```

```python
#Vérification des performances du pipeline
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

y_pred_pipe = pipeline_lr.predict(X_test)

print("Accuracy :", accuracy_score(y_test, y_pred_pipe))
print("\nMatrice de confusion :\n", confusion_matrix(y_test, y_pred_pipe))
print("\nClassification report :\n", classification_report(y_test, y_pred_pipe))
```
